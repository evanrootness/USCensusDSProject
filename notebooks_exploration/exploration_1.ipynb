# Data Exploration

```python
print(dfIndustry.shape, dfFinance.shape, dfEducation.shape)
```

```python
dfEducation['numCd'] = dfEducation['cd'].str.extract('(\d+)', expand=False).astype(int)
dfEducation['strCd'] = dfEducation['cd'].str.extract('([A-Z][A-Z])', expand=False).astype(str)

dfFinance['numCd'] = dfFinance['cd'].str.extract('(\d+)', expand=False).astype(int)
dfFinance['strCd'] = dfFinance['cd'].str.extract('([A-Z][A-Z])', expand=False).astype(str)

dfIndustry['numCd'] = dfIndustry['cd'].str.extract('(\d+)', expand=False).astype(int)
dfIndustry['strCd'] = dfIndustry['cd'].str.extract('([A-Z][A-Z])', expand=False).astype(str)
```

```python
dfEducation = dfEducation.where(dfEducation['Year'] == 2021)
dfFinance = dfFinance.where(dfFinance['Year'] == 2021)
dfIndustry = dfIndustry.where(dfIndustry['Year'] == 2021)
```

```python
dfEducation.dropna(axis = 0, how = 'any', inplace = True)
dfFinance.dropna(axis = 0, how = 'any', inplace = True)
dfIndustry.dropna(axis = 0, how = 'any', inplace = True)
```

```python
merged_data = dfEducation.merge(dfFinance, left_on = 'cd', right_on = 'cd', suffixes = ('_education', "_finance"))
all_data = merged_data.merge(dfIndustry, left_on = 'cd', right_on = 'cd')
```
